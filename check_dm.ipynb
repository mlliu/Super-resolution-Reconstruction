```python
import matplotlib.pyplot as plt
import pydicom
import pydicom.data
import nibabel as nib
```

```python
# Full path of the DICOM file is passed in base
base = r"../ARIC/scans/6_TOF_3D_5-slab_MIP_COR/DICOM/"
pass_dicom = "1.3.12.2.1107.5.2.36.40247.2018072613123013714235735"  # file name is 1-12.dcm
  
# enter DICOM image name for pattern
# result is a list of 1 element
#filename = pydicom.data.data_manager.get_files(base, pass_dicom)
filename = base +pass_dicom
ds = pydicom.dcmread(filename)
  
plt.imshow(ds.pixel_array, cmap=plt.cm.bone)  # set the color map to bone
plt.show()
```
